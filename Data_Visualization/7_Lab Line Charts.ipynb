### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here

df_comp = df[["Age", "ConvertedCompYearly"]].copy()
df_comp["ConvertedCompYearly"] = pd.to_numeric(df_comp["ConvertedCompYearly"], errors="coerce")
df_comp = df_comp.dropna()


df_comp = df_comp[df_comp["ConvertedCompYearly"] < df_comp["ConvertedCompYearly"].quantile(0.99)]

age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
age_order = [a for a in age_order if a in df_comp["Age"].unique()]

grouped = df_comp.groupby("Age")["ConvertedCompYearly"].median().reindex(age_order)
#.reindex() ensures the line goes from youngest to oldest left to right

fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(grouped.index, grouped.values,
        color="#4C72B0", linewidth=2.5,
        marker="o", markersize=8,
        markerfacecolor="white", markeredgewidth=2)

# Add value labels above each point
#Value labels are annotated above each point with ax.annotate() 
#so the exact median compensation is readable directly on the chart without hovering
for x, y in zip(grouped.index, grouped.values):
    ax.annotate(f"${y:,.0f}", xy=(x, y),
                xytext=(0, 12), textcoords="offset points",
                ha="center", fontsize=9,
                fontweight="bold", color="#4C72B0")

ax.set_title("Median Yearly Compensation by Age Group",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Age Group", fontsize=12)
ax.set_ylabel("Median Compensation", fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
plt.xticks(rotation=30, ha="right")
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
df_comp = df[["Age", "ConvertedCompYearly"]].copy()
df_comp["ConvertedCompYearly"] = pd.to_numeric(df_comp["ConvertedCompYearly"], errors="coerce")
df_comp = df_comp.dropna()


df_comp = df_comp[df_comp["ConvertedCompYearly"] < df_comp["ConvertedCompYearly"].quantile(0.99)]

age_filter = ["25-34 years old", "35-44 years old"]
df_comp = df_comp[df_comp["Age"].isin(age_filter)]

grouped = df_comp.groupby("Age")["ConvertedCompYearly"].median().reindex(age_filter)

fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(grouped.index, grouped.values,
        color="#4C72B0", linewidth=2.5,
        marker="o", markersize=8,
        markerfacecolor="white", markeredgewidth=2)

for x, y in zip(grouped.index, grouped.values):
    ax.annotate(f"${y:,.0f}", xy=(x, y),
                xytext=(0, 12), textcoords="offset points",
                ha="center", fontsize=9,
                fontweight="bold", color="#4C72B0")

ax.set_title("Median Yearly Compensation by Age Group",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Age Group", fontsize=12)
ax.set_ylabel("Median Compensation", fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
plt.xticks(rotation=30, ha="right")
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python

df_exp = df[["YearsCodePro", "JobSatPoints_6"]].copy()
df_exp["JobSatPoints_6"] = pd.to_numeric(df_exp["JobSatPoints_6"], errors="coerce")

#Replacing text entries then convert to numeric
df_exp["YearsCodePro"] = df_exp["YearsCodePro"].replace({
    "Less than 1 year": 0,
    "More than 50 years": 51
})

df_exp["YearsCodePro"] = pd.to_numeric(df_exp["YearsCodePro"], errors="coerce")
df_exp = df_exp.dropna()

#Bining experience so it does makes more sence with economic categories
bins = [0, 2, 5, 10, 15, 20, 30, 51]
labels = ["0-2", "3-5", "6-10", "11-15", "16-20", "21-30", "31+"]
df_exp["ExperienceLevel"] = pd.cut(df_exp["YearsCodePro"], bins=bins, labels=labels, right=True)

grouped = df_exp.groupby("ExperienceLevel", observed=True)["JobSatPoints_6"].median()


fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(grouped.index.astype(str), grouped.values,
        color="#4C72B0", linewidth=2.5,
        marker="o", markersize=8,
        markerfacecolor="white", markeredgewidth=2)

# Add value labels above each point
for x, y in zip(grouped.index.astype(str), grouped.values):
    ax.annotate(f"{y:.1f}", xy=(x, y),
                xytext=(0, 12), textcoords="offset points",
                ha="center", fontsize=10,
                fontweight="bold", color="#4C72B0")

ax.set_title("Median Job Satisfaction by Years of Professional Coding Experience",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Years of Experience", fontsize=12)
ax.set_ylabel("Median Job Satisfaction", fontsize=12)
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

df_exp = df[["WorkExp", "JobSatPoints_6"]].copy()
df_exp["JobSatPoints_6"] = pd.to_numeric(df_exp["JobSatPoints_6"], errors="coerce")

median_WorkExp = df["WorkExp"].median()
df["WorkExp"] = df["WorkExp"].fillna(median_WorkExp)

mean_JobSatPoints_6 = df["JobSatPoints_6"].mean()
df["JobSatPoints_6"] = df["JobSatPoints_6"].fillna(mean_JobSatPoints_6)

#Bining experience so it does makes more sence with economic categories
bins = [0, 2, 5, 10, 15, 20, 30, 51]
labels = ["0-2", "3-5", "6-10", "11-15", "16-20", "21-30", "31+"]
df_exp["WorkExp"] = pd.cut(df_exp["WorkExp"], bins=bins, labels=labels, right=True)

grouped = df_exp.groupby("WorkExp", observed=True)["JobSatPoints_6"].median()


fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(grouped.index.astype(str), grouped.values,
        color="#4C72B0", linewidth=2.5,
        marker="o", markersize=8,
        markerfacecolor="white", markeredgewidth=2)

# Add value labels above each point
for x, y in zip(grouped.index.astype(str), grouped.values):
    ax.annotate(f"{y:.1f}", xy=(x, y),
                xytext=(0, 12), textcoords="offset points",
                ha="center", fontsize=10,
                fontweight="bold", color="#4C72B0")

ax.set_title("Median Job Satisfaction by Years of Professional Coding Experience",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Years of Experience", fontsize=12)
ax.set_ylabel("Median Job Satisfaction", fontsize=12)
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```
