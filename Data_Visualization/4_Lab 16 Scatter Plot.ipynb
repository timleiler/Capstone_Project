#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

df["Age_numeric"] = df["Age"].map(age_mapping)

df = df.dropna(subset=["Age", "JobSatPoints_6"])

plt.figure(figsize=(12, 7))
sns.stripplot(data=df, x="Age_numeric", y="JobSatPoints_6",
              order="Age_numeric", alpha=0.3, s=4, color="#4C72B0")

sns.regplot(data=df, x="Age_numeric", y="JobSatPoints_6")

plt.title("Age vs. Job Satisfaction")
plt.xlabel("Age")
#plt.ylabel("Job Satisfaction")
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
print(df["Age"].unique())
print(df["YearsCodePro"].unique())
```

```python
## Write your code here

age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

df["Age_numeric"] = df["Age"].map(age_mapping)

#YearsCodePro has values like "more than 50" - we need to change that to numeric values
df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0,
    "More than 50 years": 51
})
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors="coerce")
df = df.dropna(subset=["YearsCodePro"])

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="Age_numeric", y="YearsCodePro",
            scatter_kws={"alpha": 0.3, "s": 15, "color": "#4C72B0"},
            line_kws={"color": "#DD8452", "linewidth": 2})
plt.title("Age vs. Years of Professional Coding Experience")
plt.xlabel("Age")
#plt.ylabel("Years of Professional Coding Experience")
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

df["Age_numeric"] = df["Age"].map(age_mapping)


df["ConvertedCompYearly"] = pd.to_numeric(df["ConvertedCompYearly"], errors="coerce")
df["JobSatPoints_6"] = pd.to_numeric(df["JobSatPoints_6"], errors="coerce")

df = df[df["ConvertedCompYearly"] < df["ConvertedCompYearly"].quantile(0.99)]


bubble_size = (df["Age_numeric"] / df["Age_numeric"].max()) * 100  # scale factor for visible bubbles

fig, ax = plt.subplots(figsize=(12, 7))

scatter = ax.scatter(df["ConvertedCompYearly"], df["JobSatPoints_6"],
                     s=bubble_size, alpha=0.6, color="#4C72B0",
                     edgecolors="none")

for age in [16, 21, 29, 39, 49, 59, 70]:
    ax.scatter([], [], s=age * 2, alpha=0.6, color="#4C72B0",
               edgecolors="none", label=f"Age {age}")


ax.legend(title="Age_numeric", labelspacing=1.2, frameon=True, fontsize=9)

ax.set_title("Compensation vs. Job Satisfaction\n(bubble size = Age_numeric)",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Yearly Compensation", fontsize=12)
ax.set_ylabel("Job Satisfaction", fontsize=12)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))

plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here

# LanguageHaveWorkedWith contains multiple values separated by ";"
df = df.assign(LanguageHaveWorkedWith=df["LanguageHaveWorkedWith"].str.split(";")).explode("LanguageHaveWorkedWith")
df["LanguageHaveWorkedWith"] = df["LanguageHaveWorkedWith"].str.strip()

# Getting top 10 most popular languages
top10 = df["LanguageHaveWorkedWith"].value_counts().nlargest(10).index
df = df[df["LanguageHaveWorkedWith"].isin(top10)]

# Calculate mean job satisfaction per language and sort
lang_stats = df.groupby("LanguageHaveWorkedWith")["JobSatPoints_6"].mean().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(lang_stats.values, lang_stats.index,
           s=100, color="#4C72B0", edgecolors="white", linewidth=0.5, zorder=3)

# Add horizontal grid lines for readability
ax.xaxis.grid(True, linestyle="--", alpha=0.7)
ax.set_axisbelow(True)

ax.set_title("Top 10 Programming Languages vs. Mean Job Satisfaction",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Mean Job Satisfaction")
ax.set_ylabel("Programming Language")

plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here


# Convert JobSatPoints_6 to numeric
df["JobSatPoints_6"] = pd.to_numeric(df["JobSatPoints_6"], errors="coerce")
df = df.dropna(subset=["JobSatPoints_6"])

#Contains multiple values separated by ";"
df = df.assign(LanguageHaveWorkedWith=df["LanguageHaveWorkedWith"].str.split(";")).explode("LanguageHaveWorkedWith")
df["LanguageHaveWorkedWith"] = df["LanguageHaveWorkedWith"].str.strip()

#Getting top 10 most popular languages
top10 = df["LanguageHaveWorkedWith"].value_counts().nlargest(10).index
df = df[df["LanguageHaveWorkedWith"].isin(top10)]

#Calculating mean job satisfaction per language and sort
lang_stats = df.groupby("LanguageHaveWorkedWith")["JobSatPoints_6"].mean().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(12, 7))

ax.scatter(lang_stats.values, lang_stats.index,
           s=100, color="#4C72B0", edgecolors="white", linewidth=0.5, zorder=3)

# Add horizontal grid lines for readability
ax.xaxis.grid(True, linestyle="--", alpha=0.7)
ax.set_axisbelow(True)

ax.set_title("Top 10 Programming Languages vs. Mean Job Satisfaction",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Mean Job Satisfaction", fontsize=12)
ax.set_ylabel("Programming Language", fontsize=12)

plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here

age_mapping = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

df["Age_numeric"] = df["Age"].map(age_mapping)


df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0,
    "More than 50 years": 51
})
df["YearsCodePro"] = pd.to_numeric(df["YearsCodePro"], errors="coerce")


# Age: extract first number from strings like "25-34 years old"
df["Age_numeric"] = df["Age_numeric"].str.extract(r"(\d+)").astype(float)


# Top 5 countries by frequency for color coding
top5_countries = df["Country"].value_counts().nlargest(5).index
df = df[df["Country"].isin(top5_countries)]

# Top 5 age groups by frequency for marker differentiation
top5_ages = df["Age_numeric"].value_counts().nlargest(5).index
df = df[df["Age_numeric"].isin(top5_ages)]

# Define markers for each age group
markers = ["o", "s", "D", "^", "P"]
age_groups = sorted(df["Age_numeric"].unique())
marker_map = dict(zip(age_groups, markers))

plt.figure(figsize=(14, 8))

for age, marker in marker_map.items():
    subset = df[df["Age_numeric"] == age]
    sns.scatterplot(data=subset, x="Age_numeric", y="YearsCodePro",
                    hue="Country", palette="Set2",
                    marker=marker, s=80, alpha=0.6,
                    edgecolors="none", legend=(age == age_groups[0]))

# Build custom legend for markers (age groups)
for age, marker in marker_map.items():
    plt.scatter([], [], marker=marker, color="gray",
                label=f"Age {int(age)}", s=60)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles=handles, labels=labels,
           title="Country / Age Group",
           bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=9)

plt.title("Work Experience vs. Age Group by Country",
          fontsize=14, fontweight="bold")
plt.xlabel("Age_numeric", fontsize=12)
#plt.ylabel("Years of Professional Coding Experience", fontsize=12)
plt.tight_layout()
plt.show()
```
