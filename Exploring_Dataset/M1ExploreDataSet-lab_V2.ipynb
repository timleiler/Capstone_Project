```python
## Write your code here

#Age variable = str -> I cannot combine a str with .mean()

age_mapping = {
    "Under 18 years old":    16,
    "18-24 years old":       21,
    "25-34 years old":       30,
    "35-44 years old":       40,
    "45-54 years old":       50,
    "55-64 years old":       60,
    "65 years or older":     70
}

# New colum with mean ages
df["age_mapping"] = df["Age"].map(age_mapping)

# Calculate mean
print(df["age_mapping"].mean())

```
