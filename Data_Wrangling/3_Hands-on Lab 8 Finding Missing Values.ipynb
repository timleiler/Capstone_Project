### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here

print(df.isnull().sum())
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here

plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(),            #converts the entire dataframe into True (missing) and False (not missing)
            cbar=False,             # no color bar needed, it's just True/False
            yticklabels=False,      # hide row numbers, too many to display
            cmap="viridis")       # yellow = missing, purple = not missing
            
plt.title("Missing Values Heatmap", fontsize=14, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

            
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here

df["Employment"].isnull().sum()
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here

print(df["Employment"].mode()[0])  # Find the most frequent value in a column
print("----")
print(df["Employment"].value_counts().head()) # See the top 5 most frequent values
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here

most_frequent = df["Employment"].mode()[0]
df["Employment"] = df["Employment"].fillna(most_frequent)

df["Employment"].isnull().sum()
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here

counts = df["Employment"].value_counts()
plt.figure(figsize=(10, 7))
plt.pie(counts.values, 
        labels=counts.index, 
        autopct="%1.1f%%")
plt.title("Employment Status")
plt.tight_layout()
#plt.legend()
plt.show()
```
